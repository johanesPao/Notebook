# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# disabling warning copying a slice from a dataframe
pd.options.mode.chained_assignment = None  # default='warn'
pd.options.display.float_format = '{:,.2f}'.format
```

# Reading data

```python
data = pd.read_csv('Sales2022.csv')
```

# Selecting and Cleanning Data

```python
# Selecting specific data
selected_data = data[['Art', 'Quantity', 'Bruto', 'Net', 'Cogs', 'Profit', 'brand', 'Id Cust', 'date']]

# Cleaning data and converting string to float for some columns
selected_data['Quantity'] = selected_data['Quantity'].str.replace(',','').astype(float)
selected_data['Bruto'] = selected_data['Bruto'].str.replace(',','').astype(float)
selected_data['Net'] = selected_data['Net'].str.replace(',','').astype(float)
selected_data['Cogs'] = selected_data['Cogs'].str.replace(',','').astype(float)
selected_data['Profit'] = selected_data['Profit'].str.replace(',','').astype(float)
```

## Discount Calculation

$
\begin{align}
disc = \frac{bruto - net}{bruto}
\end{align}
$

```python
selected_data['discount'] = (selected_data['Bruto'] - selected_data['Net']) / selected_data['Bruto']
```
